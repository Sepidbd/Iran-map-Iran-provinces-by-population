```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
iran=gpd.read_file('IRN_adm1.shp')
caspaian = gpd.read_file('worldglwd1.shp')
iho= gpd.read_file('iho.shp')
eez_iho= gpd.read_file('eez_iho.shp')
```

```python
fig, ax = plt.subplots(figsize=(15,15))
iran.plot(ax=ax, color='black', edgecolor='white', linewidth=0.2) # 2
caspaian.plot(ax=ax, color='#A1E2E6',linewidth=1)
iho.plot(ax=ax, color='#A1E2E6',linewidth=1)
eez_iho.plot(ax=ax, color='#A1E2E6',linewidth=1)

plt.title('Map of Iran', fontsize=25, fontname='Palatino Linotype', color='black') # 3
ax.axis('off')
plt.axis('equal')
plt.show()
fig.savefig('Map of Iran.png', dpi=300)
```

```python
df = pd.read_csv("provinces pop.csv")
df
```

```python
merged = iran.set_index('NAME_1').join(df.set_index('province'))

merged.plot()
```

```python
variable = 'population(2023)'
vmin, vmax = 1,15
fig, ax = plt.subplots(figsize=(10, 6))
merged.plot(column=variable,cmap='YlOrRd', linewidth=0.8,ax=ax, edgecolor='0.2')
ax.axis('off')
plt.title('Iran Provinces by population', fontsize=20, fontname='Palatino Linotype', color='black') 
sm = plt.cm.ScalarMappable(cmap='YlOrRd', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm, orientation='vertical')

# this will save the figure as a high-res png. you can also save as svg
fig.savefig('Iran Provinces By Population.png', dpi=300)
```
